```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
X, y=datasets.make_regression(n_samples=100, n_features=1,noise=20, random_state=4)
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=1234)
print(X_train.shape,X_test.shape)
```

```python
from linear_regression import LinearRegression

model=LinearRegression(lr=0.01, num_iter=500)
model.fit(X_train,y_train)
```

```python
y_preds=model.predict(X_test)
y_preds
```

```python
def MSE(y_preds, y_true):
    mse=np.mean((y_preds-y_true)**2)
    return mse

mse=MSE(y_preds, y_test)
mse
```

```python
cmap=plt.get_cmap('viridis')

fig=plt.figure(figsize=(6,6))
m1=plt.scatter(X_test,y_test, color=cmap(0.9), s=20 )
plt.plot(X_test,y_preds, color='black', linewidth=2, label='prediction')
```

```python
from sklearn.linear_model import LinearRegression as LR

model_2=LR()
model_2.fit(X_train,y_train)
y_preds2=model_2.predict(X_test)
mse2=MSE(y_preds2, y_test)
mse2
```
